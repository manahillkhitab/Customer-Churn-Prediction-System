```python
for col in df.select_dtypes(include='object').columns:
    print(col, "->", df[col].unique())
```

```python
cols_to_fix = [
    'MultipleLines',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
]

for col in cols_to_fix:
    df[col] = df[col].replace({
        'No internet service': 'No',
        'No phone service': 'No'
    })
```

```python
df = df.drop('customerID', axis=1)
```
